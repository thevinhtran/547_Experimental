```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(8, 6), dpi=80)
```

```python
df = pd.read_csv(r'/Users/vinhtran/Downloads/Combined.csv', header = None)
data = pd.DataFrame.to_numpy(df)
for i in range(31,len(data[:,0])):
    data[i,0] = data[i,0] - data[31,1]
for i in range(31):
    data[i,0] = data[i,0] + data[31,1]
```

```python
B = np.linspace(-3,3,50);
C = np.linspace(-0.00002,0.00005,50);
A = np.zeros(50);
```

```python
plt.plot(data[:,0], data[:,1])
plt.plot(B,A, '--', color = 'black')
plt.plot(A,C, '--', color = 'black')
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
#plt.grid()
plt.xlabel("Voltage V in Volts")
plt.ylabel("Current I in Amps")
plt.grid()
plt.title("IV Characteristics Au-Si Room Temperature")
plt.savefig('plot-2.pdf')
```
